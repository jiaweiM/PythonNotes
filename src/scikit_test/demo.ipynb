```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

style.use('ggplot')
from sklearn import svm
```

```python
x = [1, 3, 8, 1.3, 7, 6]
y = [5, 8, 3.4, 2.5, 0.1, 10]

plt.scatter(x, y)
plt.show()
```

```python
X  = np.array([
    [1, 5], 
    [3, 8], 
    [8, 3.4], 
    [1.3, 2.5], 
    [7, 0.1], 
    [6, 10]
])
t = np.array([0, 0, 1, 0, 1, 1])
```

```python
clf = svm.SVC(kernel='linear')
```

```python
clf.fit(X, t)
```

```python
clf.predict(np.array([[0.38, 0.8]]))
```
